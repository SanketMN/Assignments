## Creating a pivot table with members and the animes watched by them ##

```python
pivot = pd.pivot(data=df, index = 'members',columns='name', values='rating')
```

```python
pivot
```

## Replacing the null values ##

```python
pivot.fillna(0,inplace=True)
```

```python
pivot
```

```python
sim = cosine_similarity(pivot)
sim
```

```python
similarity = pd.DataFrame(sim)
similarity
```

```python
similarity.index = df['members'].unique()
similarity.columns = df['members'].unique()
```

```python
similarity
```

```python
def recommend_animes(members, ratings_matrix, members_similarity, top_n=5):
    similar_members = members_similarity.loc[members][members_similarity.loc[members] == 1].index.tolist()
    
    recommendations = ratings_matrix.loc[similar_members].mean(axis=0)  #Calculate the average rating for each anime across the similar members.
    recommendations = recommendations[ratings_matrix.loc[members] == 0]  #Filter out animes that the target member has already rated.

    #Sort animes by their average ratings in descending order and return the top n anime IDs.
    return recommendations.sort_values(ascending=False).head(top_n).index.tolist()
```

```python
recommend_animes(5551,pivot,similarity,5)
```

""" Question1 :- User-Based Collaborative Filtering (UBCF)
UBCF focuses on finding similar users to generate recommendations.
User similarity calculation: Calculate the similarity between users based on their past behavior (e.g., ratings, purchases).
Neighborhood formation: Identify a set of similar users (the "neighborhood") for each user.
Recommendation generation: Generate recommendations for a user by aggregating the preferences of their similar users.
Item-Based Collaborative Filtering (IBCF)
IBCF focuses on finding similar items to generate recommendations. The algorithm works as follows:
Item similarity calculation: Calculate the similarity between items based on their past interactions (e.g., co-purchases, co-ratings).
Item neighborhood formation: Identify a set of similar items for each item.
Recommendation generation: Generate recommendations for a user by finding items that are similar to the ones they have liked or interacted with."""

""" Question2 :- Collaborative Filtering
Collaborative filtering is a technique that predicts a user's preferences by analyzing the behavior of similar users. The idea is that if a group of users have similar preferences, they will likely have similar opinions about other items.
How Does Collaborative Filtering Work?
The CF process involves the following steps:
Data Collection: Gather user-item interaction data, such as ratings, purchases, or clicks.
User-Item Matrix: Create a matrix where rows represent users, columns represent items, and cell values represent the user's interaction with the item (e.g., rating or 0/1 for presence/absence of interaction).
Similarity Calculation: Calculate the similarity between users or items using metrics such as cosine similarity,Pearson correlation.
Neighborhood Formation: Identify a set of similar users (user-based CF) or items (item-based CF) for each user or item.
Recommendation Generation: Generate recommendations for a user by aggregating the preferences of their similar users (user-based CF) or by finding items that are similar to the ones they have liked or interacted with (item-based CF)."""
