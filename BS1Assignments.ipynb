## Converting categorical columns in to Numerical column using one hot encoding method.

```python
"""To put any data in to an algorithm the categorical columns need to be converted to a binary code i.e 0 and 1
as machine does not understand words or language expressions. To do this we can use the following get_dummies function from Pandas library.
It creates as many columns as the number of unique values in a categorical column and assigns a 0 or 1 value to each row"""
```

```python
pd.get_dummies(data = df, columns=['BU'], dtype=int)
```

## in the above example I have converted the BU categorical column in to three numerical column.
